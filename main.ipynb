```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data= open('ex1data1.txt',)
train_x,train_y = [], []
for line in data:
    x,y = map(float,line.strip().split(','))
    train_x.append(x)
    train_y.append(y)
```

```python
def cost(train_x, train_y, theta_1, theta_0):
    m = len(train_y)
    sum_error = 0
    for j in range(0,m):
        x = train_x[j]
        y = train_y[j]
        h_theta  = theta_1 * x + theta_0
        error  = (h_theta - y)**2
        sum_error += error
    return sum_error/(2*m)
```

```python
learning_rate = 0.01
iteration = 2000
```

```python
def gradient_descent(train_x, train_y, theta_1, theta_0, iteration):
    m = len(train_y)
    loss_history = []
    for i in range(0, iteration):
        for j in range(0, m):
            x = train_x[j]
            y = train_y[j]
            h_theta = theta_1 * x + theta_0
            error = h_theta - y 
            theta_1 = theta_1 - (1/m) * learning_rate * error * x
            theta_0   = theta_0   - (1/m) * learning_rate * error
        loss_history.append(cost(train_x, train_y, theta_1, theta_0))
    return theta_1, theta_0, loss_history
```

```python
theta_1 = 0.0 
theta_0 = 0.0
```

```python
theta_1, theta_0, loss_history = gradient_descent(train_x, train_y, theta_1, theta_0, iteration)
```

```python
m = len(train_y) 
result = []
for i in range(m):
    result.append(train_x[i]*theta_1 + theta_0)

plt.scatter(train_x, train_y)
plt.plot(train_x, result, c = 'r')
plt.title('Linear Regression Figure',fontsize=20)
plt.xlabel('Population',fontsize=20)
plt.ylabel('Profit',fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.plot(loss_history)
plt.ylabel('Loss',fontsize=20)
plt.xlabel('Iterations', fontsize=20);
plt.show()
```
